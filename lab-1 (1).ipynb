```python
!nvidia-smi
from numba import cuda
import numpy as np
@cuda.jit
def hello_kernel():
    # Thread and block indices
    tx = cuda.threadIdx.x
    bx = cuda.blockIdx.x
    bdim = cuda.blockDim.x

    # Global thread ID
    gid = bx * bdim + tx

    print("Hello from Block", bx, "Thread", tx, "Global ID", gid)
blocks = 2
threads_per_block = 4

hello_kernel[blocks, threads_per_block]()
cuda.synchronize()
```
